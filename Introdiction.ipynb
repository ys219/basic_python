#### Windows

Hello Windows users, for installations, you need to go to [this page](https://www.python.org/downloads/windows/) and download the latest version python.

Once you have done, double click the installer and follow the recommanded installation options. The latest python version should come with pip.

To verify python and pip were installated, you can check [this page](https://phoenixnap.com/kb/how-to-install-python-3-windows) for the details. 

If you pip was not installed, check [this](https://phoenixnap.com/kb/install-pip-windows) to install manually.

Module is a python file which containing defined functions, classes and variables. If you know R, it is basically equivalent to packages. Python3 comes with quite a lot of commonly used modules. But you might still need to install other modules for different purposes, pip will helps you install modules with enter the following code on cmd.

```python
pip install module_name
```

#### Apple Macintosh

Hello Mac users, here we will recommand install python with [Homebrew](https://brew.sh/) the package manager which you would also use for installing git. Of course if you feel any other way (e.g. conda, easy_install) would be more useful, you can easily found tutorial online. However, try to ensure you have only used one paltform to install python.

Install Homebrew with entering this to the terminal:

```python
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install.sh)"
```

Then you can simply do 

```python
brew install python3
```

Module is a python file which containing defined functions, classes and variables. If you know R, it is basically equivalent to packages. Python3 comes with quite a lot of commonly used modules. But you might still need to install other modules for different purposes, pip will helps you install modules. First let's check if you have installed python and pip with

```python
python3 --version 
pip3 --version
```

If you got them installed, you can now install modules with 

```python
pip3 install module_name
```

#### Ubuntu

Hello, welcome this installation instruction for ubuntu user.
First, you please check your ubuntu system version with 

```python
lsb_release -a
```

If you have ubuntu 16.04 or later versions(except 16.10 and 17.04), congratulations, python2.7 and python3.xx should be already installed, check it with

```python
python3 --version
```

If you have Ubuntu 16.10 and 17.04，simpliy install python 3.6 or later version with

```python
sudo apt-get update
sudo apt-get install python3.6
```

If you are using any ubuntu versions that are not included above, you need to first install PPA from [deadsnake](https://launchpad.net/~deadsnakes/+archive/ubuntu/ppa), as it is not aviliable in the Universe respository.

```python
sudo add-apt-repository ppa:deadsnakes/ppa
sudo apt-get update
sudo apt-get install pyhton3.6
```

Module is a python file which containing defined functions, classes and variables. If you know R, it is basically equivalent to packages. Python3 comes with quite a lot of commonly used modules.But you might still need to install other modules for different purposes, it can be usually done with

```python
sudo apt install python3-packagename
```

But certain packages need to be installed with [pip](https://pip.pypa.io/en/stable/), a python packages manager, you can install pip with

```python
sudo apt install python3-pip
```

After this, you can install packages via pip with

```python
pip3 install module_name
```

IDE stand for Intergrated Development Environments, it allows programmer edit source code, automate the configurations for several utilities and helps debuging etc. Rstudio is an IDE for R. There are also similar choice for python.

If you are looking for a Rstudio-like IDE desifned for python. Pycharm might be your choice. It can be easily installed with their [page](https://www.jetbrains.com/pycharm/)
Spyder is another python-sepcific IDE. It is a simple one with lots of tooltips and templates. [Here](https://www.spyder-ide.org/) is the official webpage with installation instructions
